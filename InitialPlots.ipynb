```python
import os
import numpy as np
import pandas as pd
import pypsa
import matplotlib.pyplot as plt

# Set the directory containing your .nc files
network_folder = 'networks/'

# Define a list of colors for each year (e.g., 2017, 2018, 2019)
colors = ['tab:blue', 'tab:orange', 'tab:green']
# Define line styles for different data types
line_styles = {'total_costs': '-', 'total_load': '--', 'renewable_generation': '-.'}

# Initialize a plot with twin y-axes
fig, ax1 = plt.subplots(figsize=(14, 7))

# Create a twin y-axis to plot load and renewable generation (MW)
ax2 = ax1.twinx()

# Load each network and extract nodal costs, total load, and renewable generation
for i, filename in enumerate(sorted(os.listdir(network_folder))):
    if filename.endswith('.nc'):
        network_path = os.path.join(network_folder, filename)
        n = pypsa.Network(network_path)
        
        # Calculate nodal cost (marginal price * load)
        nodal_costs = n.buses_t["marginal_price"] * n.loads_t["p_set"]
        total_costs = nodal_costs.sum(axis="columns")

        # Calculate total load and renewable generation
        #renewables = n.generators[n.generators.carrier.isin(["solar", "wind"])]
        renewable_generation = n.generators_t.p.sum(axis=1) #n.generators_t.p[renewables.index].sum(axis=1)
        total_load = n.loads_t.p.sum(axis=1)
        
        # Adjust the time index for each year
        total_costs.index = total_costs.index + pd.DateOffset(years=1-i)
        renewable_generation.index = renewable_generation.index + pd.DateOffset(years=1-i)
        total_load.index = total_load.index + pd.DateOffset(years=1-i)

        # Smooth the total load and renewable generation with a 7-day moving average
        smoothed_total_load = total_load.rolling(window='7D').mean()
        smoothed_renewable_generation = renewable_generation.rolling(window='7D').mean()

        # Plot nodal costs on the first y-axis (EUR) with the current year's color
        ax1.plot(total_costs.index, total_costs, label=f'{filename[:4]} - Nodal Costs (EUR)', 
                 color=colors[i], linestyle=line_styles['total_costs'])
        ax1.set_ylabel('Nodal Costs (EUR)', color='tab:blue')
        ax1.tick_params(axis='y', labelcolor='tab:blue')

        # Plot total load, renewable generation, and net load (MW) on ax2 with different line styles
        ax2.plot(smoothed_total_load.index, smoothed_total_load, label=f'{filename[:4]} - Total Load (MW)', 
                 color=colors[i], linestyle=line_styles['total_load'])
        ax2.plot(smoothed_renewable_generation.index, smoothed_renewable_generation, 
                 label=f'{filename[:4]} - Renewable Gen (MW)', color=colors[i], linestyle=line_styles['renewable_generation'])
        
        ax2.set_ylabel('Load / Renewable Generation (MW)', color='tab:green')
        ax2.tick_params(axis='y', labelcolor='tab:green')

# Set the x-axis labels to show only months
month_labels = total_costs.index.strftime('%b')
plt.xticks(ticks=total_costs.index[::744], labels=month_labels[::744], rotation=45)  # Adjust the frequency as needed

# Set the title and labels for the x-axis
plt.title('Nodal Costs, Total Load, and Renewable Generation Over Three Years')
ax1.set_xlabel('Month')

# Add legends for both axes
ax1.legend(loc='upper left')
ax2.legend(loc='upper right')

# Enable the grid for readability
ax1.grid()

# Adjust the layout to avoid overlap
plt.tight_layout()

# Show the plot
plt.show()
```

```python
import os
import numpy as np
import pandas as pd
import pypsa
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
from matplotlib.dates import date2num

# Set the directory containing your .nc files
network_folder = 'networks/outputs/solved/UC'

type_mapping = {
    '_UC': 'unchanged',
    '_MC': 'mix changed',
    '_LT': 'low transmission',
    '_HT': 'high transmission',
    '_LS': 'low storage',
    '_HS': 'high storage'
}
difficult_periods = pd.read_csv(f"{network_folder}/difficult_periods.csv")
netload_events = pd.read_csv(f"{network_folder}/netload_events.csv")

for i, filename in enumerate(sorted(os.listdir(network_folder))):
    if filename.endswith('.nc'):

        # Define the period: October to May
        period = slice(f'{filename[0:4]}-10-01', f'{int(filename[0:4])+1}-03-31')

        network_path = os.path.join(network_folder, filename)
        n = pypsa.Network(network_path)

        difficult_period = difficult_periods.loc[difficult_periods["file_name"] == filename, ["start", "end"]]
        netload_event = netload_events.loc[netload_events["file_name"] == filename, ["start", "end"]]

        # Initialize subplots (1 subplot)
        fig, axs = plt.subplots(figsize=(12, 4), sharex=False)

        # Calculate nodal cost (marginal price * load)
        nodal_costs = n.buses_t["marginal_price"] * n.loads_t["p_set"]
        total_costs = nodal_costs.sum(axis="columns").loc[period]

        # Calculate mismatch (Load - Renewable Generation)
        mismatch = n.loads_t.p.sum(axis=1).loc[period] - n.generators_t.p.sum(axis=1).loc[period]

        # Calculate accumulated flows across all lines and links
        line_flows = n.lines_t.p0.abs().sum(axis=1).loc[period]
        link_flows = n.links_t.p0.abs().sum(axis=1).loc[period]
        accumulated_flows = line_flows + link_flows
        
        # Calculate total storage level
        total_storage_level = n.storage_units_t.state_of_charge.sum(axis=1).loc[period]

        # Smooth the data with a 1-day moving average
        smoothed_mismatch = mismatch.rolling(window='1D').mean()
        smoothed_accumulated_flows = accumulated_flows.rolling(window='1D').mean()
        smoothed_total_costs = total_costs.rolling(window='1D').mean()
        smoothed_total_storage_level = total_storage_level.rolling(window='1D').mean()
        
        # Create twin axes for the current subplot
        ax1 = axs
        ax2 = ax1.twinx()
        ax3 = ax1.twinx()
        ax3.spines['right'].set_position(('outward', 55))  # Offset the third axis
        ax4 = ax1.twinx()
        ax4.spines['right'].set_position(('outward', 65))  # Offset the fourth axis
        
        # Plot nodal costs on the first y-axis (EUR)
        ax1.plot(smoothed_total_costs, label=f'Nodal Costs (EUR)', color='blue', linestyle='-')
        ax1.set_ylabel('Nodal Costs (EUR)', color='blue')
        ax1.tick_params(axis='y', labelcolor='blue')

        # Plot total load and renewable generation on the second y-axis (MW)
        ax2.plot(smoothed_mismatch, label=f'Total Mismatch (MW)', color='tab:purple', linestyle='-', alpha=0.5)
        ax2.set_ylabel('Mismatch (MW)', color='purple')
        ax2.tick_params(axis='y', labelcolor='purple')
        ax2.ticklabel_format(axis='y', style='sci', scilimits=(0, 5))  # Apply scientific notation
        
        # Plot total storage level on the third y-axis
        ax3.plot(smoothed_total_storage_level, label=f'Storage Level (MWh)', color='orange', linestyle=':')
        ax3.set_ylabel('Storage Level (MWh)', color='orange')
        ax3.tick_params(axis='y', labelcolor='orange')
        ax3.ticklabel_format(axis='y', style='sci', scilimits=(0, 7))
        ax3.get_yaxis().get_offset_text().set_position((1.1,0))

        # Plot total storage level on the third y-axis
        ax4.plot(smoothed_accumulated_flows, label=f'Acc. Flows (MWh)', color='yellow', linestyle='--')
        ax4.set_ylabel('Acc. Flows (MWh)', color='yellow')
        ax4.tick_params(axis='y', labelcolor='yellow')
        ax4.ticklabel_format(axis='y', style='sci', scilimits=(0, 7))
        ax4.get_yaxis().get_offset_text().set_position((1.1,0))

        # Extract type and name information
        for key, value in type_mapping.items():
            if key in filename:
                type = value
                name = f'{filename[0:4]} {type}'
                shortname = f'{filename[0:4]}{key}'
                break

        # Set the title for each subplot
        ax1.set_title(f'{name}')

        # Adjust x-ticks for the filtered period
        month_labels = smoothed_total_costs.index.strftime('%b')
        plt.xticks(ticks=smoothed_total_costs.index[::744], labels=month_labels[::744], rotation=45)

        # Highlight difficult periods and netload events
        if not difficult_period.empty:
            ax1.axvspan(date2num(pd.Timestamp(difficult_period.iloc[0, 0])),
                        date2num(pd.Timestamp(difficult_period.iloc[0, 1])),
                        label="extreme event", color="red", alpha=0.1)
        if not netload_event.empty:
            ax1.axvspan(date2num(pd.Timestamp(netload_event.iloc[0, 0])),
                        date2num(pd.Timestamp(netload_event.iloc[0, 1])),
                        0, 0.045, label="netload event", color="red", alpha=0.3)

        # Add legends for all axes
        ax1.legend(loc='upper left', ncol=1)
        ax2.legend(loc='upper right')
        ax3.legend(loc='lower right')
        ax4.legend(loc='lower right')

        # Enable grid and adjust layout
        ax1.grid()
        ax1.margins(x=0)
        plt.tight_layout()

        # Save the plot
        plt.savefig(f'figures/timeseries_focused_{shortname}.png')
```

```python
import os
import numpy as np
import pandas as pd
import pypsa
import matplotlib.pyplot as plt
import seaborn as sns

# Set the directory containing your .nc files
network_folder = 'networks/sample/solved'

# Load auxiliary CSVs
difficult_periods = pd.read_csv(f"{network_folder}/difficult_periods.csv")
netload_events = pd.read_csv(f"{network_folder}/netload_events.csv")

# Initialize a list to collect data for multiple years
all_costs = []

# Process each network file
for i, filename in enumerate(sorted(os.listdir(network_folder))):
    if filename.endswith('.nc'):
        network_path = os.path.join(network_folder, filename)
        n = pypsa.Network(network_path)

        # Calculate total costs
        nodal_costs = n.buses_t["marginal_price"] * n.loads_t["p_set"]
        total_costs = nodal_costs.sum(axis="columns")
        total_costs = total_costs.rolling(window='1D').mean()

        # Extract timestamp and cost
        data = pd.DataFrame({
            "timestamp": total_costs.index,
            "cost": total_costs.values
        })

        # Convert timestamp to datetime
        data["timestamp"] = pd.to_datetime(data["timestamp"])

        # Define "fiscal year" July-to-June
        data["month"] = data["timestamp"].dt.month
        data["fiscal_year"] = np.where(
            data["month"] >= 7,
            data["timestamp"].dt.year,
            data["timestamp"].dt.year - 1
        )
        data["dayofyear"] = (data["timestamp"] - pd.to_datetime(data["fiscal_year"].astype(str) + "-07-01")).dt.days

        # Add custom year label for fiscal year
        data["fiscal_year_label"] = data["fiscal_year"].astype(str) + "/" + (data["fiscal_year"] + 1).astype(str)

        # Append to the list
        all_costs.append(data)

# Combine all years into a single DataFrame
combined_data = pd.concat(all_costs)

# Pivot data for heatmap
heatmap_data = combined_data.pivot_table(
    index="fiscal_year_label",
    columns="dayofyear",
    values="cost",
    aggfunc="mean"
)

# Plot heatmap
plt.figure(figsize=(20, 10))
sns.heatmap(
    heatmap_data,
    #vmin=1,       # Minimum value for the color bar
    #vmax=10000,   # Maximum value for the color bar
    cmap="YlOrBr",
    cbar_kws={'label': 'Cost'},
    linewidths=0.5
)

# Customize x-axis to show months from July to June
xticks = pd.date_range("2022-07-01", "2023-06-30", freq="MS")
xticks_labels = xticks.strftime("%b")  # Format month names
plt.xticks(
    ticks=np.linspace(0, len(heatmap_data.columns) - 1, len(xticks_labels)),
    labels=xticks_labels,
    rotation=45
)

# Add labels and title
plt.title("Yearly Costs Heatmap (July to June)")
plt.xlabel("Month")
plt.ylabel("Fiscal Year")
plt.tight_layout()
plt.show()
```

```python
import os
import numpy as np
import pandas as pd
import pypsa
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.colors import LogNorm
from datetime import datetime
from matplotlib.dates import date2num

# Set the directory containing your .nc files
network_folder = 'networks/sample/solved'

# Load auxiliary CSVs
difficult_periods = pd.read_csv(f"{network_folder}/difficult_periods.csv")
netload_events = pd.read_csv(f"{network_folder}/netload_events.csv")

# Generate the date range for consistent x-axis ticks
num_days = 366  # Leap year (adjust if needed)
start_date = "2022-07-01"
dates = pd.date_range(start=start_date, periods=num_days, freq="D")
october_start = dates.get_loc("2022-10-01")
april_end = dates.get_loc("2023-04-30")
xticks = pd.date_range(start="2022-10-01", end="2023-04-30", freq="MS")
xtick_labels = xticks.strftime("%b")

# Initialize the figure for all heatmaps
num_years = sum(1 for f in os.listdir(network_folder) if f.endswith('.nc'))
fig, axes = plt.subplots(num_years, 1, figsize=(8, 0.8*num_years), sharex=True, constrained_layout=True)

# If there is only one subplot, ensure axes is iterable
if num_years == 1:
    axes = [axes]

# Process and plot each year inside the loop
for i, filename in enumerate(sorted(os.listdir(network_folder))):
    if filename.endswith('.nc'):
        network_path = os.path.join(network_folder, filename)
        n = pypsa.Network(network_path)

        difficult_period = difficult_periods.loc[difficult_periods["file_name"] == filename, ["start", "end"]]
        netload_event = netload_events.loc[netload_events["file_name"] == filename, ["start", "end"]]

        # Calculate total costs
        nodal_costs = n.buses_t["marginal_price"] * n.loads_t["p_set"]
        total_costs = nodal_costs.sum(axis="columns")
        total_costs = total_costs.rolling(window='1D').mean()

        # Extract timestamp and cost
        data = pd.DataFrame({
            "timestamp": total_costs.index,
            "cost": total_costs.values
        })

        # Convert timestamp to datetime
        data["timestamp"] = pd.to_datetime(data["timestamp"])

        # Define "fiscal year" July-to-June
        data["month"] = data["timestamp"].dt.month
        data["fiscal_year"] = np.where(
            data["month"] >= 7,
            data["timestamp"].dt.year,
            data["timestamp"].dt.year - 1
        )
        data["dayofyear"] = (data["timestamp"] - pd.to_datetime(data["fiscal_year"].astype(str) + "-07-01")).dt.days

        # Pivot data for heatmap
        heatmap_data = data.pivot_table(
            index="fiscal_year",
            columns="dayofyear",
            values="cost",
            aggfunc="mean"
        )

        # Plot heatmap for this year
        sns.heatmap(
            heatmap_data,
            ax=axes[i],
            norm=LogNorm(vmin=1e6, vmax=1e10),
            cmap="OrRd",
            cbar=(i == num_years - 1),
            cbar_kws={'label': 'Cost', 'orientation': 'horizontal', 'shrink': 0.5, 'aspect': 60},
            linewidths=0
        )

        # Format the subplot
        fiscal_year_label = f"{data['fiscal_year'].iloc[0]}/{data['fiscal_year'].iloc[0] + 1}"
        #axes[i].set_title(f"Year {fiscal_year_label}")
        axes[i].set_xlim(october_start, april_end)
        axes[i].set_xticks(np.linspace(october_start, april_end, len(xticks)))
        axes[i].set_xticklabels(xtick_labels, rotation=0)

        # Hide the x-axis ticks for all except the last subplot
        if i != num_years - 1:
            axes[i].tick_params(axis='x', which='both', length=0)  # Remove x-axis ticks
            axes[i].set_xticklabels([])  # Remove x-axis labels
        # Hide y-axis label for clarity
        axes[i].set_ylabel("")
        axes[i].set_xlabel("")

        if not difficult_period.empty:
            start_dayofyear = (
                pd.Timestamp(difficult_period.iloc[0, 0]).tz_localize(None) - 
                pd.Timestamp(f"{data['fiscal_year'].iloc[0]}-07-01")
            ).days
            end_dayofyear = (
                pd.Timestamp(difficult_period.iloc[0, 1]).tz_localize(None) - 
                pd.Timestamp(f"{data['fiscal_year'].iloc[0]}-07-01")
            ).days
        
            axes[i].axvspan(start_dayofyear, end_dayofyear, 0, 0.075, label="Difficult Period", color="black", alpha=0.9)
        else:
            print(f"No difficult period found for file: {filename}")

        if not netload_event.empty:
            start_dayofyear = (
                pd.Timestamp(netload_event.iloc[0, 0]).tz_localize(None) - 
                pd.Timestamp(f"{data['fiscal_year'].iloc[0]}-07-01")
            ).days
            end_dayofyear = (
                pd.Timestamp(netload_event.iloc[0, 1]).tz_localize(None) - 
                pd.Timestamp(f"{data['fiscal_year'].iloc[0]}-07-01")
            ).days
        
            axes[i].axvspan(start_dayofyear, end_dayofyear, ymin=0.075, ymax=0.15, label="Netload Event", color="blue", alpha=0.9)
        else:
            print(f"No netload events found for file: {filename}")

# Show the plot
plt.show()
```
